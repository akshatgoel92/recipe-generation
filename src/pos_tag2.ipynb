```python
# Import packages
import os
import spacy
import joblib
import pandas as pd
```

```python
def get_pipeline(model_name = "en_core_web_sm"):
    '''
    Get pipeline object
    '''
    nlp = spacy.load(model_name)
    
    return(nlp)
```

```python
def load_data(path):
    '''
    Input: Path where data is stored
    Output: Loaded data
    '''
    df = pd.read_csv(path)
    
    return(df)
```

```python
def process_recipes(df):
    '''
    Input:
    Output: 
    '''
    recipes = [text.split("', \'")[1:-1] for text in df.values.tolist()]
    
    return(recipes)
```

```python
def process_ids(df):
    '''
    Input:
    Output:
    '''
    ids = df.values.tolist()
    
    return(ids)
```
