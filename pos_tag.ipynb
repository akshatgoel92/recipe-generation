```python
# Import packages
import spacy
import pandas as pd

nlp = spacy.load("en_core_web_sm")
```

```python
def load_data(path):
    '''
    Input: Path where data is stored
    Output: Loaded data
    '''
    df = pd.read_csv(path)
    
    return(df)
```

```python
def process_recipes(df):
    '''
    Input:
    Output: 
    '''
    recipes = [text.split("', \'")[1:-1] for text in df.values.tolist()]
    
    return(recipes)
```

```python
def get_verbs(text):
    '''
    Input: Recipe [List]
    Output: Tagged Verbs [List]
    '''
    tokens = nlp(text)
    
    tags = [(token.text, token.pos_) for token in tokens]
    
    return(tags)
```

```python
def main():
    '''
    Input: None
    Output: Execute
    '''
    df = load_data('./data/RAW_recipes.csv')['steps']
    recipes = process_recipes(df)
    
    tags = []
    total_recipes = len(recipes)
    print(len(recipes))
    
    for i, recipe in enumerate(recipes):
        
        print(i/total_recipes)
        tags.append([get_verbs(text) for text in recipe])
    
    return(tags)
```

```python
tags = main()
```
